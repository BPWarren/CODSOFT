### Data Processing

```python
# Drop model 'no-need' features
dataset.columns
```

```python
# Drop useless features & Convert string values into numeric
dataset = dataset.drop(columns=['PassengerId', 'Name', 'Cabin', 'Ticket', 'Age_bins'], axis=1)
dataset['Sex']=dataset['Sex'].replace({'male':1,'female':0})
dataset.head()
```

```python
dataset.columns
```

```python
# Embarked
dataset["Embarked"].unique()
```

```python
dataset["Embarked"] = dataset["Embarked"].replace({'S':1, 'C':2, 'Q':3})
dataset.head()
```

### Handle missing values

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].median())
dataset["Age"].isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
# Manage Embarked missing values
dataset["Embarked"].mode()
dataset["Embarked"] = dataset["Embarked"].fillna(dataset["Embarked"].mode()[0]) # Because mode() returns a series
dataset.isnull().sum()
```

### Visualize correlation map

```python
corr_matrix = dataset.corr()
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

### Features & Target

```python
# Features
features = dataset.drop(columns=['Survived'])
features.head()
```

```python
# Target
target = dataset['Survived']
target.tail()
```

```python
# Normalisation
scaler = MinMaxScaler()
features = scaler.fit_transform(features)
```

```python
# reshape target
target = target.values
```

```python
target = target.reshape(-1, 1)
```

### Split data

```python
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2,shuffle=True,random_state=42)
```

### Decision tree classifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_train, y_train)
```

```python
y_predict = dtc.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
```
